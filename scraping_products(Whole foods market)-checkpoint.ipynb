```python
import selenium
from selenium import webdriver as wb
webD = wb.Chrome('chromedriver.exe')
webD.get('https://products.wholefoodsmarket.com/search?sort=relevance')
```

```python
clickobj = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[1]/div[3]/div[5]/div[2]')
clickobj.click()
```

```python
clickobj2 = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[1]/div[3]/div[4]/div[2]')
clickobj2.click()
```

```python
Productinfolist = webD.find_elements_by_class_name('ProductCard-Root--3g5WI')
```

```python
listoflinks = []
for el in Productinfolist:
    pp1 = el.find_element_by_tag_name('a')
    listoflinks.append(pp1.get_property('href'))
    
```

```python
from tqdm import tqdm

all_data = []
for i in tqdm(listoflinks):
    webD.get(i)
    #webD.find_element_by_xpath('//*[@id="app"]')
    webD.find_element_by_class_name('Collapsible-Root--3cwwH')
    webD.find_element_by_tag_name('a')
    webD.find_element_by_tag_name('h2').click()
    Ingredients = webD.find_element_by_xpath('//*[@id="ingredients-panel"]/div[1]').text
    Nameofproduct = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[2]/h1').text
    webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[2]/div[6]').click()
    Nutrionalfacts_protein = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[1]/div[1]').text
    Nutrionalfacts_omg = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[2]').text
    Nutrionalfacts_fiber = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[3]').text
    Nutrionalfacts_sugars = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[4]').text
    Servingsize = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[1]').text
    calories = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[5]').text
    
    Totalfat = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[8]/div[2]/div').text
    Totalfat_DV = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[8]/div[1]').text
    
    Saturatedfat = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[9]/div[2]/div').text
    Saturatedfat_DV = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[9]/div[1]').text
    
    Cholestrol = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[11]/div[2]/div').text
    Cholestrol_DV = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[11]/div[1]').text
    
    Sodium = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[12]/div[2]/div').text
    Sodium_DV = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[12]/div[1]').text
    
    Totalcarbohydrates = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[13]/div[2]').text
    Totalcarbohydrates_DV = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[13]/div[1]').text
    
    VitaminD =  webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[18]/div[2]/div').text
    VitaminD_DV =  webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[18]/div[1]').text
    
    Calcium = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[19]/div[2]').text
    Calcium_DV = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[19]/div[1]').text
    
   
    temp = {'Ingredients': Ingredients,
         'Nameofproduct':Nameofproduct, 
          'Nutritional facts_protein':Nutrionalfacts_protein,
            'Nutritional facts_omg':Nutrionalfacts_omg,
            'Nutritional facts_fibers':Nutrionalfacts_fiber,
            'Nutritional facts_sugars':Nutrionalfacts_sugars,
            'Servingsize':Servingsize,
             'calories':calories,
             'Totalfat':Totalfat,
              'Totalfat_DV':Totalfat_DV,
             'Saturatedfat':Saturatedfat,
              'Saturatedfat_DV':Saturatedfat_DV,
              'Cholestrol':Cholestrol,
              'Cholestrol_DV':Cholestrol_DV,
              'Sodium':Sodium,
              'Sodium_DV':Sodium_DV,
              'Totalcarbohydrates':Totalcarbohydrates,
              'Totalcarbohydrates_DV':Totalcarbohydrates_DV,
              'VitaminD':VitaminD,
              'VitaminD_DV':VitaminD_DV,
              'Calcium':Calcium,
              'Calcium_DV':Calcium_DV,
           'linkofproduct':i}
    all_data.append(temp)
    
    
```
