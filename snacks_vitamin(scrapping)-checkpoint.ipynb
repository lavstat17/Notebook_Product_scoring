```python
Productinfolist = webD.find_elements_by_class_name('filtered-item')
```

```python
listoflinks = []
for el in Productinfolist:
    pp1 = el.find_element_by_tag_name('a')
    listoflinks.append(pp1.get_property('href'))
```

```python
from tqdm import tqdm
from selenium.common.exceptions import NoSuchElementException

all_data = []
for i in tqdm(listoflinks):
    webD.get(i)
    webD.find_elements_by_class_name('product__title')
    webD.find_element_by_tag_name('p')
    nameofproduct = webD.find_element_by_xpath('//*[@id="__layout"]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[1]/p[1]').text
    webD.find_elements_by_class_name('shell')
    webD.find_element_by_xpath('//*[@id="__layout"]/div/div[2]/div[2]/div[3]/div/div[1]/section/div/nav/ul/li[2]').click()
    ingredients = webD.find_element_by_xpath('//*[@id="ingredients"]/div/div[2]/div/ul').text
    webD.find_element_by_xpath('//*[@id="__layout"]/div/div[2]/div[2]/div[3]/div/div[1]/section/div/nav/ul/li[3]').click()
      
    protein_gram = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/div[2]/div[3]/div[2]/div[1]/p[3]/span').text
    sodium_mg = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/div[2]/div[12]/div[2]/div[1]/p[3]').text
    fiber_gram = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/div[2]/div[19]/div[2]/div[1]/p[3]').text
    sugars_gram = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/div[2]/div[1]/div[2]/div[1]/p[3]').text
        
    energy = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/div[2]/div[4]/div[2]/div[1]/p[3]').text
    saturatedfat = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/div[2]/div[18]/div[2]/div[1]/p[3]').text
    servingsize = webD.find_element_by_xpath('//*[@id="section-info"]/div/div[2]/div/div[2]/p').text

   
    temp = {
            'nameofproduct':nameofproduct, 
            'ingredients': ingredients,
            'protein_gram':protein_gram,
            'sodium_mg':sodium_mg,
            'fiber_gram':fiber_gram,
            'sugars_gram':sugars_gram,
            'energy':energy,
             'saturatedfat':saturatedfat,
            'servingsize':servingsize,
           'linkofproduct':i}
    all_data.append(temp)
    
```

```python
import pandas as pd
pd.DataFrame(all_data)
```
