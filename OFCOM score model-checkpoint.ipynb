```python
#convert each serving size to 100g
df_new = df[['protein_g','Saturated_Fat_g','sugars_g','Servingsize_g','sodium_mg','fiber_g','Energy/calories']].astype(int)
d = {30:3.34, 28:3.58, 55:1.83, 40:2.5, 26:3.85, 26:3.85, 40:2.5, 48:2.1, 35:2.86}
df_new['constant'] = df['Servingsize_g'].map(d)
df_new['protein_100g'] = df_new['protein_g']*df_new['constant'].astype(int)
df_new['Saturated_Fat_100g'] = df_new['Saturated_Fat_g']*df_new['constant'].astype(int)
df_new['sugars_100g'] = df_new['sugars_g']*df_new['constant']
df_new['fiber_100g'] = df_new['fiber_g']*df_new['constant']
df_new['Servingsize_100g'] = df_new['Servingsize_g']*df_new['constant']
df_new['Energy_KJ'] = df_new['Energy/calories']*4.184
df_new['Energy_KJ'] = df_new['Energy_KJ'].astype(int)
```

```python
# Defining negative factors 
def negative_factor(row):
    if row['Energy_KJ'] < 335 and row['Energy_KJ'] ==335 or row['Saturated_Fat_100g'] <1 and row['Saturated_Fat_100g'] ==1 or row['sugars_100g'] <4.5 and row['sugars_100g'] ==4.5 or row['sodium_mg'] < 90 and row['sodium_mg'] ==90:
        return 0
    elif row['Energy_KJ'] >335 or row['Saturated_Fat_100g'] >1 or row['sugars_100g'] > 4.5 or row['sodium_mg'] > 90:
        return 1
    elif row['Energy_KJ'] >670 or row['Saturated_Fat_100g'] >2  or row['sugars_100g'] > 9 or row['sodium_mg'] > 180:
        return 2
    elif row['Energy_KJ'] >1005 or row['Saturated_Fat_100g'] >3 or row['sugars_100g'] >13.5 or row['sodium_mg'] >270:
        return 3
    elif row['Energy_KJ'] >1340 or row['Saturated_Fat_100g'] >4 or row['sugars_100g'] >18 or row['sodium_mg'] > 360:
        return 4
    elif row['Energy_KJ'] >1675 or row['Saturated_Fat_100g'] >5 or row['sugars_100g'] > 22.5 or row['sodium_mg'] >450:
        return 5
    elif row['Energy_KJ'] >2010 or row['Saturated_Fat_100g'] >6 or row['sugars_100g'] > 27 or row['sodium_mg'] > 540:
        return 6
    elif row['Energy_KJ'] >2345 or row['Saturated_Fat_100g'] >7 or row['sugars_100g'] > 31 or row['sodium_mg'] > 630:
        return 7
    elif row['Energy_KJ'] >2680 or row['Saturated_Fat_100g'] >8 or row['sugars_100g'] > 36 or row['sodium_mg'] > 720:
        return 8
    elif row['Energy_KJ'] >3015 or row['Saturated_Fat_100g'] >9 or row['sugars_100g'] > 40 or row['sodium_mg'] > 810:
        return 9
    elif row['Energy_KJ'] >3350 or row['Saturated_Fat_100g'] >10 or row['sugars_100g'] > 45 or row['sodium_mg'] > 900:
        return 10
```

```python
#Defining positive factors
def positive_factor(row):
    if row['Fruit_Veg_Nuts_percentage'] <40 and row['Fruit_Veg_Nuts_percentage'] ==40 or row['NSP_Fibre_g'] <0.7 and row['NSP_Fibre_g'] ==0.7 or row['AOAC_Fibre_g'] <0.9 and row['AOAC_Fibre_g'] ==0.9 or row['Protein_g'] <1.6 and row['Protein_g'] ==1.6:
        return 0
    elif row['Fruit_Veg_Nuts_percentage'] >40 or row['NSP_Fibre_g'] >0.7 or row['AOAC_Fibre_g'] >0.9 or row['Protein_g'] >1.6:
        return 1
    elif row['Fruit_Veg_Nuts_percentage'] >60 or row['NSP_Fibre_g'] >1.4 or row['AOAC_Fibre_g'] >1.9 or row['Protein_g'] >3.2:
        return 2
    elif row['NSP_Fibre_g'] >2.1 or row['AOAC_Fibre_g'] >2.8 or row['Protein_g'] >4.8:
        return 3
    elif row['NSP_Fibre_g'] >2.8 or row['AOAC_Fibre_g'] >3.7 or row['Protein_g'] >6.4:
        return 4
    elif row['Fruit_Veg_Nuts_percentage'] >80 or row['NSP_Fibre_g'] >3.5 or row['AOAC_Fibre_g'] > 4.7 or row['Protein_g'] >8.0:
        return 5
```

```python
df_new1['energy_score'] = df_new1.apply(lambda row : negative_factor(row),axis=1)
df_new1['sodium_score'] = df_new1.apply(lambda row : negative_factor(row),axis=1)
df_new1
```
