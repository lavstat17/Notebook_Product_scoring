```python
#scrapping the organic products from wholefoods markets website - snacks(chips,candy,cookies) category
```

```python
import selenium
from selenium import webdriver as wb
webD = wb.Chrome('chromedriver.exe')
webD.get('https://products.wholefoodsmarket.com/search?sort=relevance')
```

```python
clickobj = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[1]/div[3]/div[6]')
clickobj.click()
```

```python
clickobj1 = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[1]/div[3]/div[10]/div[1]')
clickobj1.click()
```

```python
clickobj2 = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[1]/div[3]/div[3]/div[1]')
clickobj2.click()
```

```python
clickobj3 = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[1]/div[4]/div[5]/div[2]')
clickobj3.click()
```

```python
Productinfolist = webD.find_elements_by_class_name('ProductCard-Root--3g5WI')
```

```python
listoflinks = []
for el in Productinfolist:
    pp1 = el.find_element_by_tag_name('a')
    listoflinks.append(pp1.get_property('href'))
    
```

```python
listoflinks
```

```python
len(listoflinks)
```

```python
from tqdm import tqdm
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By

all_data = []
for i in tqdm(listoflinks):
    webD.get(i)
    try:
        webD.find_elements_by_class_name('Collapsible-Root--3cwwH')
        webD.find_element_by_tag_name('a')
        webD.find_element_by_tag_name('h2').click()
        Ingredients = webD.find_element_by_xpath('//*[@id="ingredients-panel"]/div[1]').text
        Nameofproduct = webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[2]/h1').text
        webD.find_element_by_xpath('//*[@id="app"]/div/div/div/div[2]/div[2]/div[6]').click()
        
        protein_gram = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[1]/div[1]').text
        sodium_mg = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[2]/div[1]').text
        fiber_gram = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[3]/div[1]').text
        sugars_gram = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[1]/div[4]/div[1]').text
    
        Servingsize = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[2]/div[3]').text
        
        calories_value = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[5]/div/div[3]').text
        webD.find_element_by_class_name('NutritionTable-Root--YcoZx')
        Saturatedfat = webD.find_element_by_xpath('//*[@id="nutrition-facts-panel"]/div[2]/div[9]/div[2]/div').text
        Saturatedfat = Saturatedfat.replace("\n", "")
    
    except NoSuchElementException:
        print('NA')
   
    temp = {'Ingredients': Ingredients,
         'Nameofproduct':Nameofproduct, 
          'protein_gram':protein_gram,
            'sodium_mg':sodium_mg,
            'fiber_gram':fiber_gram,
            'sugars_gram':sugars_gram,
            'Servingsize':Servingsize,
              'calories_value':calories_value,
             'Saturatedfat':Saturatedfat,
           'linkofproduct':i}
    all_data.append(temp)
```

```python
import pandas as pd

pd.DataFrame(all_data)
```

```python
chips_organic = pd.DataFrame(all_data)
chips_organic.to_csv('C:\\Users\\lavanya\\Downloads\\chips_organic.csv')
```
